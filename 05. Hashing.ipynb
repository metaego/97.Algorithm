# 딕셔너리

## 딕셔너리 삽입

```python
# 결론:
# hash의 인덱스에는 딕스타입빼고 다 가능
# hash의 value는 모든 타입 가능

hash = dict() # hash= {}

hash[1] = 'apple'
hash['banana'] = 3
hash[(4, 5)] = [1,2,3]
hash[10] = dict({1 : 'a', 2 :'b'})
hash[2.5] = (1,2,3)
# hash[{1 : 'hello', 2 : 'word'}] = 0   # TypeError: unhashable type: 'dict'
hash[True] = True  # True가 1이므로 index가 1로 변경되어 1 index값이 'apple'에서 True로 변경됨
hash[False] = False
```

```python
hash
```

## 딕셔너리 정렬

### 딕셔너리 오름차순 정렬

```python
hash = dict({1:10, 3:12, 5:7, 7:6, 4:5})
```

```python
hash
```

```python
# key값 추출 후 오름차순 정렬
# sorted(정렬할 데이터, key=파라미터, reverse=Fasle)  # key=파라미터: key를 기준으로 정렬 # reverse=Fasle: 기본값 
sorted(hash.keys(), key = lambda x:x)
```

```python
sorted(hash.keys(), key=lambda x:x, reverse=True)  # reverse=True: 내림차순에 의한 정렬
```

```python
# valuse값 추출 후 오름차순 정렬
sorted(hash.values(), key=lambda x:x)
```

```python
# valuse값 추출 후 내림차순 정렬
sorted(hash.values(), key=lambda x:x, reverse=True)
```

```python
# key와 values를 추출 후 키값을 기준으로 오름차순 정렬
sorted(hash.items(), key=lambda x:x)
```

```python
# key와 values를 추출 후 values값을 기준으로 오름차순 정렬
sorted(hash.items(), key=lambda x:x[1])
```

```python
# key와 values를 추출 후 키값을 기준으로 내림차순 정렬
sorted(hash.items(), key=lambda x:x, reverse=True)
```

```python
# key와 values를 추출 후 valuse값을 기준으로 내림차순 정렬
sorted(hash.items(), key=lambda x:-x[1])
```

### 딕셔너리 내림차순 정렬

```python
# 키값 추출 후 키값을 기준으로 내림차순 정렬
sorted(hash.keys(), key = lambda x: -x)
```

```python
# values값 추출 후 values값을 기준으로 내림차순 정렬
sorted(hash.values(), key = lambda x: -x)
```

```python
# 튜플의 두번째 원소에 의한 내림차순
sorted(hash.items(), key=lambda x:-x[-1])
```
