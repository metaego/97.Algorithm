# 탐욕법
- 전체가 아닌 현재 상태에서 최선의 선택을 하는 알고리즘
- 전체 탐색보다 빠르지만 반드시 정답을 도출하지 않는다


## 탐욕법의 조건
- 각 부분에서의 선택이 다른 부분에게 영향을 주지 않는다
- 각 부분에서의 최적해결이 최종 해결방법이다

## 탐욕법의 예제

### 동전의 갯수
- 동전들이 서로간 배수인 경우에 탐욕법 적용 가능
- 서로간 배수가 아니면 동적 계획법을 사용해야 함

### 가능한 많은 수의 회의 시간

### 카드 게임
- 두명의 사람이 서로의 카드를 비교하여 숫자가 큰 값을 가진 사람이 이기는 게임
