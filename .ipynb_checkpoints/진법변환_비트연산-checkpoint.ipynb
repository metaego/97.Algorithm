# 진법변환 & 비트연산

- 진법 변환: 우리가 지금 쓰고있는 10진법의 수를 2진법이나 8진법으로 바꾸는 걸 의미
- 비트 연산: 2진수에 대해 비트단위로 적용되는 연산
    - 종류: NOT, OR, XOR, AND

- 진법이란?
    - 수를 셀 때 자릿수가 올라가는 단위를 기준으로 하는 셈법의 총칭
    - 사용하는 숫자의 개수가 진법의 숫자를 의미

- 진법을 사용하는 이유?
    - 컴퓨터의 CPU: 2진법을 사용

- 진법 변환하기(10진법 -> 2진법)


```python
# 10진법 -> 2진법
# 10을 2진법으로 변환하는 코드

bin(10)  # 0b : 2진법을 의미하는 binary의 약자
```

```python
# 10진법 -> 8진법
# 95를 8진법으로 변환하는 코드

oct(95)  # 0o: 8진법을 의미하는 octal의 약자
```

# 10진법을 16진법으로 변환


```python
# 20진법 -> 16진법
# 350을 16진법으로 변환하는 코드
hex(350)  # 0x: 16진법을 의미하는 hex의 약자
```
